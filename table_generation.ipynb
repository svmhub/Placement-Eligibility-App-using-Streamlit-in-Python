### 3(iv). To fill the data on the placement table

```python
import sqlite3
import random
from faker import Faker
```

```python
conn = sqlite3.connect("D:/MANI/2025/PLACEMENT.db")
cursor = conn.cursor()
fake = Faker()
```

```python
cursor.execute("SELECT * FROM students")
student_ids = [row[0] for row in cursor.fetchall()]
```

```python
statuses = ['Ready', 'Not Ready', 'Placed']
companies = ['Google', 'Microsoft', 'Infosys', 'CTS', 'TCS', 'Capgemini', 'HCL', 'Wipro', 'Accenture']
```

```python
for student_id in student_ids:

    mock_interview_score = random.randint(40, 100)
    internships_completed = random.randint(0, 3)
    placement_status = random.choice(statuses)

    if placement_status == 'Placed':
        company_name = random.choice(companies)
        placement_package = random.randint(3000000, 15000000)
        interview_rounds_cleared = random.randint(1,5)
        placement_date = fake.date_between(start_date='-1y', end_date='today').strftime('%Y-%m-%d')
    else:
        company_name = None
        placement_package = None
        interview_rounds_cleared = None
        placement_date = None
    
    cursor.execute("""INSERT INTO placement (student_id, mock_interview_score, internships_completed, placement_status, company_name, placement_package, interview_rounds_cleared, placement_date) VALUES (?, ?, ?, ?, ?, ?, ?, ?)""", (student_id, mock_interview_score, internships_completed, placement_status, company_name, placement_package, interview_rounds_cleared, placement_date))

conn.commit()
conn.close()
```
